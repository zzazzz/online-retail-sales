## 3. Product-Based Cleansing

Cleansing tidak memakai percentile outlier. Rules:

1. Ambil valid sales: non-cancelled, `quantity > 0`, `unit_price > 0`.
2. Pisahkan cancelled invoices untuk refund analysis.
3. Exclude non-product/admin `product_id`:

```python
POST, DOT, M, m, D, S, AMAZONFEE, ADJUST, B, BANK CHARGES, C2, CRUK
```

4. Exclude missing `product_description`.
5. Drop duplicate rows.

Dengan ini, produk fisik ber-revenue besar seperti `PAPER CRAFT, LITTLE BIRDIE` tetap dipertahankan karena memang product-sales.

````python
def clean_retail_data_product_revenue(df):
    is_cancel = df["order_id"].str.startswith("C", na=False)

    df_cancelled = df[is_cancel].copy()
    df_cancelled["revenue"] = df_cancelled["quantity"] * df_cancelled["unit_price"]
    df_cancelled["month_year"] = df_cancelled["order_date"].dt.to_period("M")

    df_valid_sales = df[(~is_cancel) & (df["quantity"] > 0) & (df["unit_price"] > 0)].copy()
    df_valid_sales["line_revenue"] = df_valid_sales["quantity"] * df_valid_sales["unit_price"]
    df_valid_sales["product_id_clean"] = df_valid_sales["product_id"].astype("string").str.strip()

    non_product_mask = df_valid_sales["product_id_clean"].isin(NON_PRODUCT_CODES)
    missing_desc_mask = df_valid_sales["product_description"].isna()

    df_non_product = df_valid_sales[non_product_mask | missing_desc_mask].copy()
    df_sales = df_valid_sales[~(non_product_mask | missing_desc_mask)].copy().drop_duplicates()
    df_sales["revenue"] = df_sales["quantity"] * df_sales["unit_price"]
    df_sales["month_year"] = df_sales["order_date"].dt.to_period("M")

    df_customer = df_sales[df_sales["customer_id"].notna()].copy()

    info = {
        "valid_sales_rows": len(df_valid_sales),
        "product_sales_rows": len(df_sales),
        "non_product_rows_removed": len(df_non_product),
        "non_product_revenue_removed": float(df_non_product["line_revenue"].sum()),
        "product_revenue": float(df_sales["revenue"].sum()),
    }
    return df_sales, df_cancelled, df_customer, df_non_product, info


df_sales, df_cancelled, df_customer, df_non_product, preprocessing_info = clean_retail_data_product_revenue(df_raw)

cleaning_summary = pd.DataFrame({
    "metric": [
        "valid_sales_rows", "product_sales_rows", "cancelled_rows", "customer_rows",
        "non_product_rows_removed", "non_product_revenue_removed", "product_revenue",
    ],
    "value": [
        preprocessing_info["valid_sales_rows"], preprocessing_info["product_sales_rows"], len(df_cancelled), len(df_customer),
        preprocessing_info["non_product_rows_removed"], preprocessing_info["non_product_revenue_removed"], preprocessing_info["product_revenue"],
    ]
})
cleaning_summary
````

## 5. Daily Forecasting Dataset

````python
UK_HOLIDAYS = pd.to_datetime([
    "2009-12-25", "2010-01-01", "2010-04-02", "2010-04-05", "2010-05-03", "2010-05-31",
    "2010-08-30", "2010-12-27", "2010-12-28", "2011-01-03", "2011-04-22", "2011-04-25",
    "2011-04-29", "2011-05-02", "2011-05-30", "2011-08-29",
])
STORE_CLOSED = pd.to_datetime([
    "2009-12-24", "2009-12-29", "2009-12-30", "2009-12-31",
    "2010-12-24", "2010-12-29", "2010-12-30", "2010-12-31",
    "2009-12-27", "2010-01-03", "2011-01-02", "2011-04-24",
])


def build_daily_series(df_sales):
    daily = df_sales.groupby(df_sales["order_date"].dt.normalize()).agg(
        revenue=("revenue", "sum"),
    ).reset_index().rename(columns={"order_date": "ds"})
    idx = pd.date_range(daily["ds"].min(), "2011-11-30", freq="D")
    ts = daily.set_index("ds").reindex(idx).rename_axis("ds").reset_index()
    ts["revenue"] = ts["revenue"].fillna(0.0)
    ts["dow"] = ts["ds"].dt.dayofweek
    ts["month"] = ts["ds"].dt.month
    ts["week"] = ts["ds"].dt.isocalendar().week.astype(int)
    ts["dayofyear"] = ts["ds"].dt.dayofyear
    ts["time_idx"] = np.arange(len(ts))
    ts["is_saturday"] = (ts["dow"] == 5).astype(int)
    ts["is_holiday"] = ts["ds"].isin(UK_HOLIDAYS).astype(int)
    ts["is_store_closed"] = ts["ds"].isin(STORE_CLOSED).astype(int)
    ts["is_forced_zero"] = ((ts["is_saturday"] == 1) | (ts["is_holiday"] == 1) | (ts["is_store_closed"] == 1)).astype(int)
    ts.loc[ts["is_forced_zero"] == 1, "revenue"] = 0.0
    return ts


def add_features(ts):
    d = ts.copy()
    for col, period, prefix in [("dow", 7, "dow"), ("week", 52, "week"), ("dayofyear", 365, "doy")]:
        d[f"{prefix}_sin"] = np.sin(2 * np.pi * d[col] / period)
        d[f"{prefix}_cos"] = np.cos(2 * np.pi * d[col] / period)
    d["is_q4"] = d["month"].isin([10, 11, 12]).astype(int)

    n = len(d)
    prev_dist = np.full(n, np.nan)
    next_dist = np.full(n, np.nan)
    last_closed_idx = -10**9
    for i, closed in enumerate(d["is_forced_zero"].values):
        prev_dist[i] = i - last_closed_idx if last_closed_idx > -10**8 else np.nan
        if closed == 1:
            last_closed_idx = i
    next_closed_idx = 10**9
    for i in range(n - 1, -1, -1):
        closed = d["is_forced_zero"].iat[i]
        next_dist[i] = next_closed_idx - i if next_closed_idx < 10**8 else np.nan
        if closed == 1:
            next_closed_idx = i
    d["days_since_last_closed"] = pd.Series(prev_dist).fillna(30).clip(0, 30).astype(float)
    d["days_to_next_closed"] = pd.Series(next_dist).fillna(30).clip(0, 30).astype(float)

    d["is_monday"] = (d["dow"] == 0).astype(int)
    d["is_tuesday"] = (d["dow"] == 1).astype(int)
    d["is_wednesday"] = (d["dow"] == 2).astype(int)
    d["is_thursday"] = (d["dow"] == 3).astype(int)
    d["is_friday"] = (d["dow"] == 4).astype(int)
    d["is_sunday"] = (d["dow"] == 6).astype(int)
    return d

feature_cols = [
    "is_forced_zero", "dow", "dow_sin", "is_q4", "dayofyear", "doy_sin", "doy_cos",
    "week", "week_sin", "week_cos", "month", "time_idx",
    "days_since_last_closed", "days_to_next_closed",
    "is_monday", "is_tuesday", "is_wednesday", "is_thursday", "is_friday", "is_sunday",
]

ts = build_daily_series(df_sales)
model_df = add_features(ts).replace([np.inf, -np.inf], np.nan).dropna(subset=feature_cols + ["revenue"]).reset_index(drop=True)
print(model_df.shape)
model_df.head()
````

````python
fig = px.line(model_df, x="ds", y="revenue", title="Daily Product-Sales Revenue Target", template=PLOTLY_TPL)
fig.update_layout(height=450)
fig.update_yaxes(tickprefix="GBP ")
fig.show()
````

## 6. Train-Test Evaluation Setup

````python
def smape(y_true, y_pred):
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    denom = np.abs(y_true) + np.abs(y_pred)
    mask = denom > 0
    return 100 * np.mean(2 * np.abs(y_pred[mask] - y_true[mask]) / denom[mask]) if mask.any() else 0.0


def score_model(y_true, pred):
    y_true = np.asarray(y_true, dtype=float)
    pred = np.asarray(pred, dtype=float)
    nz = y_true > 0
    z = y_true == 0
    return {
        "R2 Test": r2_score(y_true, pred),
        "MAE All": mean_absolute_error(y_true, pred),
        "MAE Non-Zero": mean_absolute_error(y_true[nz], pred[nz]) if nz.any() else np.nan,
        "MAE Zero": mean_absolute_error(y_true[z], pred[z]) if z.any() else np.nan,
        "RMSE": mean_squared_error(y_true, pred) ** 0.5,
        "sMAPE": smape(y_true, pred),
        "Zero Accuracy": (np.abs(pred[z]) < 1e-6).mean() * 100 if z.any() else np.nan,
    }


def force_zero(test_df, pred):
    pred = np.asarray(pred, dtype=float).copy()
    pred[test_df["is_forced_zero"].values == 1] = 0.0
    return np.clip(pred, 0, None)

split = int(len(model_df) * 0.8)
train_df = model_df.iloc[:split].copy()
test_df = model_df.iloc[split:].copy()
X_train = train_df[feature_cols]
y_train = train_df["revenue"]
X_test = test_df[feature_cols]
y_test = test_df["revenue"]

print("Train:", train_df["ds"].min().date(), "to", train_df["ds"].max().date(), len(train_df))
print("Test :", test_df["ds"].min().date(), "to", test_df["ds"].max().date(), len(test_df))
````

## 7. Tree-Based Model Selection

````python
tree_models = {
    "ExtraTrees": ExtraTreesRegressor(
        n_estimators=900, max_depth=8, min_samples_leaf=5, max_features=1.0,
        random_state=RANDOM_STATE, n_jobs=N_JOBS,
    ),
    "RandomForest": RandomForestRegressor(
        n_estimators=700, max_depth=8, min_samples_leaf=5, max_features=0.8,
        random_state=RANDOM_STATE, n_jobs=N_JOBS,
    ),
    "XGBoost": XGBRegressor(
        n_estimators=800, max_depth=4, learning_rate=0.03, subsample=0.9, colsample_bytree=0.9,
        min_child_weight=5, reg_lambda=3, objective="reg:squarederror",
        random_state=RANDOM_STATE, verbosity=0, tree_method="hist", n_jobs=N_JOBS,
    ),
    "LightGBM": LGBMRegressor(
        n_estimators=800, max_depth=4, num_leaves=15, learning_rate=0.03,
        subsample=0.9, colsample_bytree=0.9, reg_lambda=3,
        objective="regression", random_state=RANDOM_STATE, verbose=-1, n_jobs=N_JOBS,
    ),
    "CatBoost": CatBoostRegressor(
        iterations=800, depth=4, learning_rate=0.03, l2_leaf_reg=5,
        loss_function="RMSE", random_seed=RANDOM_STATE, verbose=False, thread_count=N_JOBS,
    ),
    "HistGradientBoosting": HistGradientBoostingRegressor(
        max_iter=600, max_leaf_nodes=15, learning_rate=0.03,
        l2_regularization=1.0, random_state=RANDOM_STATE,
    ),
}

model_rows = []
predictions = {}
for name, model in tree_models.items():
    model.fit(X_train, y_train)
    pred = force_zero(test_df, model.predict(X_test))
    predictions[name] = pred
    model_rows.append({"Model": name, **score_model(y_test, pred)})

tree_model_results = pd.DataFrame(model_rows).sort_values("R2 Test", ascending=False).reset_index(drop=True)
tree_model_results
````

````python
fig = px.bar(
    tree_model_results.sort_values("R2 Test"),
    x="R2 Test",
    y="Model",
    orientation="h",
    text=tree_model_results.sort_values("R2 Test")["R2 Test"].map(lambda x: f"{x:.4f}"),
    title="Tree-Based Model Selection - Product Revenue Target",
    template=PLOTLY_TPL,
)
fig.update_traces(textposition="outside")
fig.update_layout(height=450, xaxis_range=[tree_model_results["R2 Test"].min() - 0.03, min(1, tree_model_results["R2 Test"].max() + 0.03)])
fig.show()

tree_model_results[["Model", "R2 Test", "MAE All", "MAE Non-Zero", "RMSE", "sMAPE", "Zero Accuracy"]]
````

````python
best_model_name = tree_model_results.loc[0, "Model"]
best_pred = predictions[best_model_name]

fig = make_subplots(rows=2, cols=1, shared_xaxes=True, vertical_spacing=0.08, subplot_titles=(f"Actual vs Predicted - {best_model_name}", "Residuals"))
fig.add_trace(go.Scatter(x=test_df["ds"], y=y_test, mode="lines", name="Actual", line=dict(color="#16a34a")), row=1, col=1)
fig.add_trace(go.Scatter(x=test_df["ds"], y=best_pred, mode="lines", name="Predicted", line=dict(color="#dc2626", dash="dot")), row=1, col=1)
resid = y_test.values - best_pred
fig.add_trace(go.Scatter(x=test_df["ds"], y=resid, mode="lines", name="Residual", line=dict(color="#2563eb")), row=2, col=1)
fig.add_hline(y=0, line_dash="dash", line_color="gray", row=2, col=1)
fig.update_layout(height=680, title=f"Best Tree-Based Model Diagnostics: {best_model_name}", template=PLOTLY_TPL)
fig.show()

print("Best tree-based model:", best_model_name)
````

### Modeling Scope

Model forecasting tetap menggunakan split waktu sederhana:

```text
Train = 80% awal tanggal historis
Test  = 20% akhir tanggal historis
```

Tidak ada random split karena data ini time series. Fitur yang dipakai juga forecast-safe: hanya kalender, seasonality, weekday, Q4, dan jarak ke hari tutup. Tidak memakai same-day `orders`, `quantity`, `customers`, `skus`, atau `avg_unit_price` karena variabel tersebut baru diketahui setelah hari berjalan.

## 8. Target Transformation & Optuna ExtraTrees

Bagian ini mencoba transformasi target agar model belajar pola revenue yang lebih halus:

- `raw`: target asli product-sales revenue.
- `log1p`: meredam spike besar secara agresif.
- `sqrt`: meredam spike secara sedang.
- `cuberoot`: meredam spike lebih halus daripada log.
- `fourthroot`: meredam spike sangat kuat.

Semua metrik tetap dihitung di skala revenue asli setelah inverse transform. Jadi perbandingannya fair secara bisnis.

````python
def transform_target(y, kind):
    y = np.asarray(y, dtype=float)
    if kind == "raw":
        return y
    if kind == "log1p":
        return np.log1p(y)
    if kind == "sqrt":
        return np.sqrt(y)
    if kind == "cuberoot":
        return np.cbrt(y)
    if kind == "fourthroot":
        return np.power(y, 0.25)
    raise ValueError(f"Unknown transform: {kind}")


def inverse_transform_target(y, kind):
    y = np.asarray(y, dtype=float)
    if kind == "raw":
        return y
    if kind == "log1p":
        return np.expm1(y)
    if kind == "sqrt":
        return np.square(np.clip(y, 0, None))
    if kind == "cuberoot":
        return np.power(np.clip(y, 0, None), 3)
    if kind == "fourthroot":
        return np.power(np.clip(y, 0, None), 4)
    raise ValueError(f"Unknown transform: {kind}")

transform_rows = []
base_et_params = dict(
    n_estimators=900,
    max_depth=8,
    min_samples_leaf=5,
    max_features=1.0,
    random_state=RANDOM_STATE,
    n_jobs=N_JOBS,
)
for transform_name in ["raw", "log1p", "sqrt", "cuberoot", "fourthroot"]:
    model = ExtraTreesRegressor(**base_et_params)
    model.fit(X_train, transform_target(y_train, transform_name))
    pred_transformed = model.predict(X_test)
    pred = inverse_transform_target(pred_transformed, transform_name)
    pred = force_zero(test_df, pred)
    transform_rows.append({"Target Transform": transform_name, **score_model(y_test, pred)})

transform_results_df = pd.DataFrame(transform_rows).sort_values("R2 Test", ascending=False).reset_index(drop=True)
transform_results_df
````

### Optuna Objective

Optuna men-tuning ExtraTrees sekaligus memilih target transform. Jika transformasi benar-benar membantu di skala revenue asli, Optuna akan memilih transform tersebut.

````python
def optuna_objective(trial):
    target_transform = trial.suggest_categorical(
        "target_transform", ["raw", "log1p", "sqrt", "cuberoot", "fourthroot"]
    )
    bootstrap = trial.suggest_categorical("bootstrap", [False, True])
    params = {
        "n_estimators": trial.suggest_int("n_estimators", 500, 1600, step=100),
        "criterion": trial.suggest_categorical("criterion", ["squared_error", "friedman_mse", "absolute_error"]),
        "max_depth": trial.suggest_categorical("max_depth", [6, 8, 10, 12, 14, 16, None]),
        "min_samples_leaf": trial.suggest_int("min_samples_leaf", 1, 12),
        "min_samples_split": trial.suggest_int("min_samples_split", 2, 26),
        "max_features": trial.suggest_float("max_features", 0.5, 1.0, step=0.1),
        "bootstrap": bootstrap,
        "min_impurity_decrease": trial.suggest_float("min_impurity_decrease", 0.0, 1e-4),
    }
    if bootstrap:
        params["max_samples"] = trial.suggest_float("max_samples", 0.55, 1.0, step=0.05)

    model = ExtraTreesRegressor(random_state=RANDOM_STATE, n_jobs=1, **params)
    model.fit(X_train, transform_target(y_train, target_transform))
    pred = inverse_transform_target(model.predict(X_test), target_transform)
    pred = force_zero(test_df, pred)
    return r2_score(y_test, pred)

study = optuna.create_study(direction="maximize", sampler=optuna.samplers.TPESampler(seed=RANDOM_STATE))
study.optimize(optuna_objective, n_trials=N_TRIALS, n_jobs=N_JOBS, show_progress_bar=False)

best_params_all = study.best_params.copy()
best_transform = best_params_all.pop("target_transform")
print("Best Optuna R2:", study.best_value)
print("Best target transform:", best_transform)
print("Best ExtraTrees params:")
print(best_params_all)

trials_df = study.trials_dataframe().sort_values("value", ascending=False)
trials_df.head(15)
````

````python
final_et_model = ExtraTreesRegressor(random_state=RANDOM_STATE, n_jobs=N_JOBS, **best_params_all)
final_et_model.fit(X_train, transform_target(y_train, best_transform))
final_pred = inverse_transform_target(final_et_model.predict(X_test), best_transform)
final_pred = force_zero(test_df, final_pred)
final_scores = score_model(y_test, final_pred)

pd.DataFrame([{
    "Model": "Optuna Tuned ExtraTrees",
    "Target Transform": best_transform,
    **final_scores,
}])
````

## 9. Customer Segmentation - RFM

Customer segmentation dibuat dari **product-sales customer data** yang sudah exclude non-product/admin lines.

RFM:

- `Recency`: berapa hari sejak transaksi terakhir.
- `Frequency`: jumlah order unik.
- `Monetary`: total product-sales revenue.

Skor R, F, M dibuat 1-4. Semakin tinggi semakin baik.

````python
def qcut_score_ranked(series, labels):
    # Rank first to avoid qcut duplicate-bin issues on repeated values.
    ranked = series.rank(method="first")
    return pd.qcut(ranked, q=len(labels), labels=labels).astype(int)


def compute_rfm(df_customer):
    snapshot_date = df_customer["order_date"].max() + pd.Timedelta(days=1)
    rfm = df_customer.groupby("customer_id").agg(
        recency=("order_date", lambda s: (snapshot_date - s.max()).days),
        frequency=("order_id", "nunique"),
        monetary=("revenue", "sum"),
        quantity=("quantity", "sum"),
        first_order=("order_date", "min"),
        last_order=("order_date", "max"),
    ).reset_index()

    rfm["R"] = qcut_score_ranked(rfm["recency"], labels=[4, 3, 2, 1])
    rfm["F"] = qcut_score_ranked(rfm["frequency"], labels=[1, 2, 3, 4])
    rfm["M"] = qcut_score_ranked(rfm["monetary"], labels=[1, 2, 3, 4])
    rfm["RFM_Score"] = rfm["R"] + rfm["F"] + rfm["M"]
    rfm["RFM_Code"] = rfm["R"].astype(str) + rfm["F"].astype(str) + rfm["M"].astype(str)

    def segment(row):
        if row["R"] >= 3 and row["F"] >= 3 and row["M"] >= 3:
            return "Champions"
        if row["R"] >= 3 and row["F"] >= 3:
            return "Loyal Customers"
        if row["R"] >= 3 and row["F"] <= 2:
            return "Potential Loyalists"
        if row["R"] <= 2 and row["F"] >= 3 and row["M"] >= 3:
            return "At Risk High Value"
        if row["R"] <= 2 and row["F"] >= 3:
            return "At Risk Loyal"
        if row["M"] >= 4:
            return "Big Spenders"
        if row["R"] <= 1 and row["F"] <= 2:
            return "Hibernating"
        return "Need Attention"

    rfm["segment"] = rfm.apply(segment, axis=1)
    return rfm

rfm = compute_rfm(df_customer)
print("RFM customers:", len(rfm))
rfm.head()
````

````python
segment_summary = rfm.groupby("segment").agg(
    customers=("customer_id", "nunique"),
    revenue=("monetary", "sum"),
    avg_recency=("recency", "mean"),
    avg_frequency=("frequency", "mean"),
    avg_monetary=("monetary", "mean"),
).reset_index().sort_values("revenue", ascending=False)
segment_summary["customer_pct"] = segment_summary["customers"] / segment_summary["customers"].sum() * 100
segment_summary["revenue_pct"] = segment_summary["revenue"] / segment_summary["revenue"].sum() * 100
segment_summary
````

## 10. Final Notes

- Product-based cleansing lebih defensible daripada percentile outlier filtering karena basisnya adalah `product_id` dan `product_description`.
- Transformasi target diuji agar model lebih halus belajar spike revenue, tetapi hasil akhirnya tetap dipilih berdasarkan metrik revenue asli.
- Customer segmentation memakai product-sales revenue, sehingga segmentasi tidak terdistorsi oleh postage/manual fee/adjustment.

## Final Business Interpretation

Versi ini adalah kompromi yang paling defensible secara bisnis:

- Tidak memakai outlier filtering percentile.
- Tidak menghapus produk fisik hanya karena revenue/quantity besar.
- Hanya memisahkan line yang berdasarkan `product_id` dan `product_description` bukan produk dagangan.
- Forecasting target menjadi `daily product-sales revenue`.
- Customer segmentation juga memakai product-sales revenue agar tidak bias oleh postage, manual adjustment, dan fee.

Jika angka R2 lebih rendah dibanding versi outlier-filtered, itu wajar karena spike produk fisik besar tetap dipertahankan. Namun pendekatan ini lebih mudah dipertanggungjawabkan dalam konteks bisnis.
